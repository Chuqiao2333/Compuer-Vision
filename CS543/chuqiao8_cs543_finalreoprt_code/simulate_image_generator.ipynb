```python
import numpy as np
import time
import os
np.random.seed(0)
```

xyz file Paramiters 

```python
#file name
xtalnm = 'WSeTe'
filesuffix = '_incostem_'
#size of a unit cell
pixel_length  = 0.2077 #A
image_size = 1500#pixel
a = 3.297#A
b = a* np.sqrt(3)
c = 5*1.2982
atomZ1 = 74#intencity
atomZ2 = 34
#initial position
atom1=np.array([atomZ1, 0.000000, 0.000000, 1.797500, 1, 0.08])
atom2=np.array([atomZ2, a/2, b/6, 0.000000, 1, 0.08])
atom3=np.array([atomZ2, a/2, b/6, 3.595000, 1, 0.08])
atom4=np.array([atomZ1, a/2, b/2, 1.797500, 1, 0.08])
atom5=np.array([atomZ2, a, b*2/3, 0.000000, 1, 0.08])
atom6=np.array([atomZ2, a, b*2/3, 3.595000, 1, 0.08])
#defect prob
dopant_conc_2Te = 0.08
dopant_conc_TeSe = 0.06
vacancy_conc_1Se = 0.03
vancancy_conc_2Se = 0.005
FileNumber = 50
```

Parameter file Parameters

```python
#Constant
#Every parameter is gaussian ditribution, the first element is mean and the second is scale
#First Line
voltage = 80#kV
Cs3_param = (0,0.001)#um
Cs5_param = (0,1)#mm
df = 0
aperture = 25.2#mrad
#Second
ADF_angle_min = 63
ADF_angle_max = 200
#Thrid
Higher_order = 'END'#Don't use for now
#Fourth
Source_size_param = (1.05,0.2)#nm
#Fifth
defocus_spread_param = (20,5)#A

counting_noise = 'y'
probe_current_param = (30,2)#A
dwell_time = 20
```

User only need to change the paramters in above cells

```python
#Other parameters determined by above
#unit cell repeat times
rep_x  = int(image_size*pixel_length/a)
rep_y = int(image_size*pixel_length/b)
rep_z = 1
#defects
z_dopant_Te = 52 #dopant atom 
z_difference_dopant = 52-34
#defect probability
dopant_conc_2Te_low = 0
dopant_conc_2Te_high = dopant_conc_2Te_low+dopant_conc_2Te
dopant_conc_TeSe_low = dopant_conc_2Te_high
dopant_conc_TeSe_high = dopant_conc_TeSe_low+dopant_conc_TeSe
vacancy_conc_1Se_low = dopant_conc_TeSe_high
vacancy_conc_1Se_high = vacancy_conc_1Se_low+vacancy_conc_1Se
vancancy_conc_2Se_low = vacancy_conc_1Se_high
vancancy_conc_2Se_high = vancancy_conc_2Se_low+vancancy_conc_2Se
#supercell size
supercell_a = a*rep_x
supercell_b = b*rep_y
supercell_c = c*rep_z
cellsizeline=(supercell_a,supercell_b,supercell_c)
```

Generate Files

```python

#One batch file for batch operate 
Batch_File_name = 'Batch_'+str(FileNumber)+'files_'+xtalnm+filesuffix+str(rep_x)+'_'+str(rep_y)+'_'+str(rep_z)+'.bat'
file_batch = open(Batch_File_name,'w+')

for N in range(FileNumber):
    
    #image files
    filename = xtalnm+filesuffix+str(rep_x)+'_'+str(rep_y)+'_'+str(rep_z)+'_'+str(N)
    fid = open(filename+'.xyz', 'w+')
    
    #Label files
    filename_2Te = xtalnm+'_2Te'+filesuffix+str(rep_x)+'_'+str(rep_y)+'_'+str(rep_z)+'_'+str(N)
    fid_2Te = open(filename_2Te+'.xyz', 'w+')
    filename_TeSe = xtalnm+'_TeSe'+filesuffix+str(rep_x)+'_'+str(rep_y)+'_'+str(rep_z)+'_'+str(N)
    fid_TeSe = open(filename_TeSe+'.xyz', 'w+')
    filename_Se = xtalnm+'_Se'+filesuffix+str(rep_x)+'_'+str(rep_y)+'_'+str(rep_z)+'_'+str(N)
    fid_Se = open(filename_Se+'.xyz', 'w+')
    filename_vacancy = xtalnm+'_vacancy'+filesuffix+str(rep_x)+'_'+str(rep_y)+'_'+str(rep_z)+'_'+str(N)
    fid_vacancy = open(filename_vacancy+'.xyz', 'w+')
    
    #Write commands into the batch file 
    batch_str = 'incostem.exe<'+filename+'.param \n'
    batch_2Te_str = 'incostem.exe<'+filename_2Te+'.param \n'
    batch_TeSe_str = 'incostem.exe<'+filename_TeSe+'.param \n'
    batch_Se_str = 'incostem.exe<'+filename_Se+'.param \n'
    batch_vacancy_str = 'incostem.exe<'+filename_vacancy+'.param \n'
    file_batch.write(batch_str)
    file_batch.write(batch_2Te_str)
    file_batch.write(batch_TeSe_str)
    file_batch.write(batch_Se_str)
    file_batch.write(batch_vacancy_str)
    
    #Write the header into each xyz file
    DateString = time.strftime("%Y_%m_%d_%H_%M_%S", time.localtime())
    xyzheader=xtalnm+filesuffix+str(rep_x)+'_'+str(rep_y)+'_'+str(rep_z)+'_Created_at_'+DateString+'\n'
    xyzheader_2Te=xtalnm+'_2Te'+filesuffix+str(rep_x)+'_'+str(rep_y)+'_'+str(rep_z)+'_Created_at_'+DateString+'\n'
    xyzheader_TeSe=xtalnm+'_TeSe'+filesuffix+str(rep_x)+'_'+str(rep_y)+'_'+str(rep_z)+'_Created_at_'+DateString+'\n'
    xyzheader_Se=xtalnm+'_Se'+filesuffix+str(rep_x)+'_'+str(rep_y)+'_'+str(rep_z)+'_Created_at_'+DateString+'\n'
    xyzheader_vacancy=xtalnm+'_vacancy'+filesuffix+str(rep_x)+'_'+str(rep_y)+'_'+str(rep_z)+'_Created_at_'+DateString+'\n'
    fid.write(xyzheader)
    fid_2Te.write(xyzheader_2Te)
    fid_TeSe.write(xyzheader_TeSe)
    fid_Se.write(xyzheader_Se)
    fid_vacancy.write(xyzheader_vacancy)
    
    # write xyz coordinates
    fid.write('{:.4f} {:.4f} {:.4f} \n'.format(cellsizeline[0],cellsizeline[1],cellsizeline[2]))
    fid_2Te.write('{:.4f} {:.4f} {:.4f} \n'.format(cellsizeline[0],cellsizeline[1],cellsizeline[2]))
    fid_TeSe.write('{:.4f} {:.4f} {:.4f} \n'.format(cellsizeline[0],cellsizeline[1],cellsizeline[2]))
    fid_Se.write('{:.4f} {:.4f} {:.4f} \n'.format(cellsizeline[0],cellsizeline[1],cellsizeline[2]))
    fid_vacancy.write('{:.4f} {:.4f} {:.4f} \n'.format(cellsizeline[0],cellsizeline[1],cellsizeline[2]))
    for i in range(rep_x):
        for j in range(rep_y):
            for k in range(rep_z):
                #seperate the first three and the last three
                #First Three
                prob1 = np.random.rand()
                single_Se1,single_Se2,vaccancy1,vaccancy2 = False,False,False,False
                if dopant_conc_2Te_low<prob1<dopant_conc_2Te_high:
                    final_atom2 = atom2+ i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])+np.array([z_difference_dopant,0,0,0,0,0])
                    fid_2Te.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom2[0],final_atom2[1],final_atom2[2],final_atom2[3],
                                                                                  final_atom2[4],final_atom2[5]))
                    final_atom3 = atom3+ i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])+np.array([z_difference_dopant,0,0,0,0,0])
                    fid_2Te.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom3[0],final_atom3[1],final_atom3[2],final_atom3[3],
                                                                                  final_atom3[4],final_atom3[5]))
                elif dopant_conc_TeSe_low <prob1<dopant_conc_TeSe_high:
                    final_atom2 = atom2+ i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])+np.array([z_difference_dopant,0,0,0,0,0])
                    fid_TeSe.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom2[0],final_atom2[1],final_atom2[2],final_atom2[3],
                                                                                  final_atom2[4],final_atom2[5]))
                    final_atom3 = atom3+ i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])
                elif vacancy_conc_1Se_low<prob1<vacancy_conc_1Se_high:
                    final_atom2 = atom2+ i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])
                    fid_Se.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom2[0],final_atom2[1],final_atom2[2],final_atom2[3],
                                                                                  final_atom2[4],final_atom2[5]))
                    final_atom3 = atom3+ i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])
                    #decide if wirte this atom or not as a vaccancy
                    single_Se1 = True
                elif vancancy_conc_2Se_low<prob1<vancancy_conc_2Se_high:
                    final_atom2 = atom2+ i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])
                    fid_vacancy.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom2[0],final_atom2[1],final_atom2[2],final_atom2[3],
                                                                                  final_atom2[4],final_atom2[5]))
                    final_atom3 = atom3+ i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])
                    fid_vacancy.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom3[0],final_atom3[1],final_atom3[2],final_atom3[3],
                                                                                  final_atom3[4],final_atom3[5]))
                    vaccancy1 = True
                else:
                    final_atom2 = atom2 + i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])
                    final_atom3 = atom3 + i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])
                final_atom1 = atom1 + i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])    
                
                
                prob2  = np.random.rand()
                if dopant_conc_2Te_low<prob2<dopant_conc_2Te_high:
                    final_atom5 = atom5+ i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])+np.array([z_difference_dopant,0,0,0,0,0])
                    fid_2Te.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom5[0],final_atom5[1],final_atom5[2],final_atom5[3],
                                                                                  final_atom5[4],final_atom5[5]))
                    final_atom6 = atom6+ i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])+np.array([z_difference_dopant,0,0,0,0,0])
                    fid_2Te.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom6[0],final_atom6[1],final_atom6[2],final_atom6[3],
                                                                                  final_atom6[4],final_atom6[5]))
                elif dopant_conc_TeSe_low<prob2<dopant_conc_TeSe_high:
                    final_atom5 = atom5+ i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])+np.array([z_difference_dopant,0,0,0,0,0])
                    fid_TeSe.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom5[0],final_atom5[1],final_atom5[2],final_atom5[3],
                                                                                  final_atom5[4],final_atom5[5]))
                    final_atom6 = atom6+ i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])
                    
                elif vacancy_conc_1Se_low<prob2<vacancy_conc_1Se_high:
                    final_atom5 = atom5+ i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])
                    fid_Se.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom5[0],final_atom5[1],final_atom5[2],final_atom5[3],
                                                                                  final_atom5[4],final_atom5[5]))
                    final_atom6 = atom6+ i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])
                    single_Se2 = True
                elif vancancy_conc_2Se_low<prob2<vancancy_conc_2Se_high:
                    final_atom5 = atom5+ i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])
                    fid_vacancy.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom5[0],final_atom5[1],final_atom5[2],final_atom5[3],
                                                                                  final_atom5[4],final_atom5[5]))
                    final_atom6 = atom6+ i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])
                    fid_vacancy.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom6[0],final_atom6[1],final_atom6[2],final_atom6[3],
                                                                                  final_atom6[4],final_atom6[5]))
                    vaccancy2 = True
                else:
                    final_atom5 = atom5 + i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])
                    final_atom6 = atom6 + i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])
                final_atom4 = atom4 + i*np.array([0,a,0,0,0,0])+j*np.array([0,0,b,0,0,0])+k*np.array([0,0,0,c,0,0])
                fid.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom1[0],final_atom1[1],final_atom1[2],final_atom1[3],
                                                                                  final_atom1[4],final_atom1[5]))
                #Not single se or vacancy, write Two atoms
                if single_Se1 == False and vaccancy1 == False:
                    fid.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom2[0],final_atom2[1],final_atom2[2],final_atom2[3],
                                                                                      final_atom2[4],final_atom2[5]))
                    fid.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom3[0],final_atom3[1],final_atom3[2],final_atom3[3],
                                                                                      final_atom3[4],final_atom3[5]))
                #Single vaccancy Wirte in one atom
                elif single_Se1 == True:
                    fid.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom2[0],final_atom2[1],final_atom2[2],final_atom2[3],
                                                                                      final_atom2[4],final_atom2[5]))
                
                elif vaccancy1 == True:
                    pass
                fid.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom4[0],final_atom4[1],final_atom4[2],final_atom4[3],
                                                                                  final_atom4[4],final_atom4[5]))
                if single_Se2 == False and vaccancy2 == False:
                    fid.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom5[0],final_atom5[1],final_atom5[2],final_atom5[3],
                                                                                      final_atom5[4],final_atom5[5]))
                    fid.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom6[0],final_atom6[1],final_atom6[2],final_atom6[3],
                                                                                      final_atom6[4],final_atom6[5]))
                elif single_Se2 == True:
                    fid.write('{:.0f} {:.6f} {:.6f} {:.6f} {:.0f} {:.2f} \n'.format(final_atom5[0],final_atom5[1],final_atom5[2],final_atom5[3],
                                                                                      final_atom5[4],final_atom6[5]))
                else:
                    pass
    
    #-1 as End of the file and close all xyz files
    fid.write('-1')
    fid_2Te.write('-1')
    fid_TeSe.write('-1')
    fid_Se.write('-1')
    fid_vacancy.write('-1')
    
    fid.close()
    fid_2Te.close()
    fid_TeSe.close()
    fid_vacancy.close()
    fid_Se.close()
    
    #Generate parameter files
    fid_param = open(filename+'.param','w+')
    fid_2Te_param = open(filename_2Te+'.param','w+')
    fid_TeSe_param = open(filename_TeSe+'.param','w+')
    fid_Se_param = open(filename_Se+'.param','w+')
    fid_vacancy_param = open(filename_vacancy+'.param','w+')
    #Wirte the xyz filename as the header
    fid_param.write(filename+'.xyz \n1 1 1\n')
    fid_2Te_param.write(filename_2Te+'.xyz \n1 1 1\n')
    fid_TeSe_param.write(filename_TeSe+'.xyz \n1 1 1\n')
    fid_Se_param.write(filename_Se+'.xyz \n1 1 1\n')
    fid_vacancy_param.write(filename_vacancy+'.xyz \n1 1 1\n')
    #Save the image as .tif files
    fid_param.write(filename+'.tif')
    fid_2Te_param.write('2Te_'+filename+'.tif')
    fid_TeSe_param.write('TeSe_'+filename+'.tif')
    fid_Se_param.write('Se_'+filename+'.tif')
    fid_vacancy_param.write('vacancy_'+filename+'.tif')
    #Set parameters
    
    image_size_str = '\n'+str(image_size)+' '+str(image_size)
    Cs3 = np.random.normal(Cs3_param[0],Cs3_param[1])
    Cs5 = np.random.normal(Cs5_param[0],Cs5_param[1])
    STEM_Param_str = '\n'+str(voltage)+' '+str(Cs3)+' '+str(Cs5)+' '+str(df)+' '+str(aperture)
    ADF_str = '\n' + str(ADF_angle_min) + ' '+ str(ADF_angle_max)
    High_order_str = '\n'+Higher_order
    Source_size = np.random.normal(Source_size_param[0],Source_size_param[1])
    source_size_str = '\n'+str(Source_size)
    defocus_spread = np.random.normal(defocus_spread_param[0],defocus_spread_param[1])
    Defocus_str = '\n'+str(defocus_spread)
    probe_current = np.random.normal(probe_current_param[0],probe_current_param[1])
    noise_str = '\n'+str(probe_current)+' '+str(dwell_time)
    
    GeneralParam = image_size_str+STEM_Param_str+ADF_str+High_order_str+source_size_str+Defocus_str+'\nn'+'\n-1'
    GeneralParam_defect = image_size_str+STEM_Param_str+ADF_str+High_order_str+source_size_str+Defocus_str+'\nn'+'\n-1'
    
    fid_param.write(GeneralParam)
    fid_2Te_param.write(GeneralParam_defect)
    fid_TeSe_param.write(GeneralParam_defect)
    fid_Se_param.write(GeneralParam_defect)
    fid_vacancy_param.write(GeneralParam_defect)
    #close param files
    fid_param.close()
    fid_2Te_param.close()
    fid_TeSe_param.close()
    fid_vacancy_param.close()
    fid_Se_param.close()
#close batch files
file_batch.close()

    
```
